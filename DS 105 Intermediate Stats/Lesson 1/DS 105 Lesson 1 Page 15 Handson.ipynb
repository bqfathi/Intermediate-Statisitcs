### Looking at this data, I want to see if age and BestSquatKg are related, as well as BestSquatKg and BodyweightKg

## Analyzing if the age and BestSquatKg are related

```python
strong['Age'].corr(strong['BestSquatKg'])
```

#### There is a weak/nonexistant negative correlation between age and BestSquatKg

## Analyzing if the BodyweightKg and BestSquatKg are related

```python
strong['BodyweightKg'].corr(strong['BestSquatKg'])
```

#### There is a weak but not nonexistant positive correlation between BodyweightKg and BestSquatKg

## Analyzing if BestBenchKg and BestBenchKg are related

```python
strong['BestBenchKg'].corr(strong['BestSquatKg'])
```

#### There is a definite correlation between BestBenchKg and BestSquatKg

## Analyzing if BestBenchKg and BestDeadlift are related

```python
strong['BestBenchKg'].corr(strong['BestDeadliftKg'])
```

#### Not as strong as I thought it would be, but still strong.
