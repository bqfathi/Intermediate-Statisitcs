## Number of trips and participants in Winter, Spring and Summer Analysis


### Number of trips Winter

```python
parks['# of trips Winter'].hist()
```

##### Skewed right (positive), but I want to try a different plot to be sure from Seaborn

```python
sns.displot(data = parks, x = '# of trips Winter', kde = True)
```

### Winter Participants

```python
sns.displot(data = parks, x = '# of participants Winter', kde = True)
```

##### Positively skewed, going to use square root transformation

```python
parks['participantsWinterSQRT'] = np.sqrt(parks['# of participants Winter'])
```
